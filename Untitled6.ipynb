```python
# Dataset Selection
from sklearn.datasets import load_wine
import pandas as pd

# Load dataset
data = load_wine()
df = pd.DataFrame(data.data, columns=data.feature_names)
df['target'] = data.target

# Preview the data
df.head()
```

```python
# Preprocessing
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.impute import SimpleImputer

# 1. Handle missing values
imputer = SimpleImputer(strategy='mean')
df_imputed = pd.DataFrame(imputer.fit_transform(df), columns=df.columns)

# 2. Feature scaling
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df_imputed.drop('target', axis=1))

# Optional: Convert to DataFrame for easier handling
X_scaled_df = pd.DataFrame(X_scaled, columns=df_imputed.columns[:-1])
```

```python
# K-Means Clustering
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt

# Determine optimal number of clusters using Elbow method
inertia = []
for k in range(2, 11):
    km = KMeans(n_clusters=k, random_state=42)
    km.fit(X_scaled_df)
    inertia.append(km.inertia_)

plt.plot(range(2,11), inertia, marker='o')
plt.title("Elbow Method")
plt.xlabel("Number of clusters")
plt.ylabel("Inertia")
plt.show()

# Apply K-Means with chosen clusters
kmeans = KMeans(n_clusters=3, random_state=42)
labels_km = kmeans.fit_predict(X_scaled_df)
silhouette_km = silhouette_score(X_scaled_df, labels_km)
print("K-Means Silhouette Score:", silhouette_km)
```

```python
# Hierarchical Clustering
from scipy.cluster.hierarchy import dendrogram, linkage, fcluster

# Create linkage matrix
Z = linkage(X_scaled_df, method='ward')

# Dendrogram
plt.figure(figsize=(10, 5))
dendrogram(Z, truncate_mode='level', p=5)
plt.title("Hierarchical Clustering Dendrogram")
plt.xlabel("Samples")
plt.ylabel("Distance")
plt.show()

# Cut the dendrogram for cluster assignment
labels_hc = fcluster(Z, t=3, criterion='maxclust')
silhouette_hc = silhouette_score(X_scaled_df, labels_hc)
print("Hierarchical Clustering Silhouette Score:", silhouette_hc)
```

```python
# Dimensionality Reduction
from sklearn.decomposition import PCA

# Reduce to 2D for visualization
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled_df)

plt.figure(figsize=(8,6))
plt.scatter(X_pca[:,0], X_pca[:,1], c=labels_km, cmap='viridis', s=50)
plt.title("K-Means Clusters Visualized with PCA")
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.show()
```

```python
from sklearn.metrics import silhouette_score, davies_bouldin_score

# K-Means
sil_km = silhouette_score(X_scaled_df, labels_km)
db_km = davies_bouldin_score(X_scaled_df, labels_km)

# Hierarchical
sil_hc = silhouette_score(X_scaled_df, labels_hc)
db_hc = davies_bouldin_score(X_scaled_df, labels_hc)

print(f"K-Means: Silhouette Score = {sil_km:.3f}, Davies-Bouldin = {db_km:.3f}")
print(f"Hierarchical: Silhouette Score = {sil_hc:.3f}, Davies-Bouldin = {db_hc:.3f}")
```
